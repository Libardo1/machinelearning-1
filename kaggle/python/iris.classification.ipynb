```python
import sklearn as sk
import sklearn.datasets
```

```python
d = sk.datasets.load_iris()
```

```python
print d.DESCR
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import pandas.tools.plotting
```

```python
df = pd.DataFrame(data=d.data,columns=d.feature_names)
print df.columns
```

```python
def plotScatterMatrix(df, classes):
    fig = plt.figure(figsize=(10, 10))

    n = len(df.columns)
    for i in range(n) :
      for j in range(n) :
        p= plt.subplot(n, n, j*n+i+1)
        if (i==j):
          p.hist(df.iloc[:,i], bins=50, color='#ddddff')
        else :
          p.scatter(df.iloc[:,i], df.iloc[:,j], c=classes, cmap='prism', alpha=0.50)

plotScatterMatrix(df, d.target)
```

```python
import sklearn.neighbors
```

```python
knn = sk.neighbors.KNeighborsClassifier(n_neighbors=3)

#this should score 1 (if no neighbours included)
knn.fit(d.data, d.target)
knn.score(d.data,d.target)
```

```python
import sklearn.cross_validation
```

```python
import numpy as np
np.random.seed(123)


knn = sk.neighbors.KNeighborsClassifier(n_neighbors=3)
kf = sk.cross_validation.KFold(len(d.data), n_folds=5, shuffle=True)

scores= []
for train_index, test_index in kf:
  knn.fit(d.data[train_index,], d.target[train_index])
  score = knn.score(d.data[test_index,], d.target[test_index])
  scores.append(score)

scores = np.asarray(scores)
print(scores)
print("Mean score = %f"%scores.mean())
```

Cross validation score using leave one out

MinMax normalization (feature wise)

```python
from sklearn import preprocessing
```

```python
scaler = sk.preprocessing.MinMaxScaler() 
data=scaler.fit_transform(d.data)
df_minmaxScaled = pd.DataFrame(data=data,columns=d.feature_names)
```

```python
df_minmaxScaled.head()
```

```python
plotScatterMatrix(df_minmaxScaled, d.target)
```

```python
def score_knn(data, target):
    np.random.seed(123)
    knn = sk.neighbors.KNeighborsClassifier(n_neighbors=5)
    kf = sk.cross_validation.KFold(len(data), n_folds=5, shuffle=True)

    scores= []
    for train_index, test_index in kf:
      knn.fit(data[train_index,], d.target[train_index])
      score = knn.score(data[test_index,], target[test_index])
      scores.append(score)
      
    scores = np.asarray(scores)
    print(scores)
    print("Mean score = %f"%scores.mean())
    

score_knn(df.values, d.target)
score_knn(df_minmaxScaled.values, d.target)
```

```python
scaler = sk.preprocessing.MinMaxScaler() 
data = sk.preprocessing.normalize(d.data, norm='l2')
data = scaler.fit_transform(data)
df_minmaxScaledNormalized = pd.DataFrame(data=data,columns=d.feature_names)
```

```python
df_minmaxScaledNormalized.head()
```

```python
plotScatterMatrix(df_minmaxScaledNormalized, d.target)
```

```python
score_knn(df.values, d.target)
score_knn(df_minmaxScaled.values, d.target)
score_knn(df_minmaxScaledNormalized.values, d.target)
```

```python
df_minmaxScaledNormalized.iloc[:,2:4].head()
```

Normalizing and scaling feature wise seems to work well with our 5-fold cross validation

```python
np.random.seed(123)

knn = sk.neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(np.asarray(df_minmaxScaledNormalized.iloc[:,2:4]), d.target)

test = []
for x in np.arange(0,1,0.001) :
  for y in np.arange(0,1,0.001) :
    test.append([x,y])

test_df = pd.DataFrame(data=test)
knn_p = pd.DataFrame(data=knn.predict_proba(test_df))   
knn_p[3] = knn.predict(test_df)
```

```python
fig = plt.figure(figsize=(20, 10))
p= plt.subplot(1, 2, 1)
p.scatter(df_minmaxScaledNormalized.iloc[:,2]*1000, df_minmaxScaledNormalized.iloc[:,3]*1000, c=d.target, s=50, cmap='prism', alpha=0.7)


p.imshow(np.reshape(knn_p[0], (1000,1000)), interpolation='nearest', cmap='Reds', alpha=0.3)
#plt.imshow(np.reshape(knn_p[1], (1000,1000)), interpolation='nearest', cmap='PRGn', alpha=0.3)
p.imshow(np.reshape(knn_p[2], (1000,1000)), interpolation='nearest', cmap='Greens', alpha=0.3)

p= plt.subplot(1, 2, 2)
p.scatter(df_minmaxScaledNormalized.iloc[:,2]*1000, df_minmaxScaledNormalized.iloc[:,3]*1000, c=d.target, s=50, cmap='prism', alpha=0.7)
p.imshow(np.reshape(knn_p[3], (1000,1000)), interpolation='nearest', alpha=0.3)
```
